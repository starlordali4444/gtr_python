### Hard Q89 — Type Casting
You have a list `["10", "20", "30"]`.  
Convert each element to an integer and print the sum.


```python
# Hard Q89 — Type Casting
values = ["10", "20", "30"]
total = 0
for v in values:
    total += int(v)
print(total)
```
